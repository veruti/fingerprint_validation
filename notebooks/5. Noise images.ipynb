## Make function for getting gradient

```python
def get_gradient(image, rx=1, ry=1, is_scharr=True):

    if is_scharr:
        dx = cv.Scharr(image, cv.CV_64F, rx, 0)
        dy = cv.Scharr(image, cv.CV_64F, 0, ry)
    else:
        dx = cv.Sobel(image, cv.CV_64F, rx, 0)
        dy = cv.Sobel(image, cv.CV_64F, 0, ry)
        
    return dx, dy
```
