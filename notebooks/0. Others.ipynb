## Make function for getting results

```python
def findContoursResult(res):
    if len(res) == 2:
        return res[0], res[1]
    elif len(res) == 3:
        return res[1], res[2]
    else:
        return None
```

```python
res = cv.findContours(inv_im, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
contours, hierarchy = findContoursResult(res)
```
